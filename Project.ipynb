# **1. Importing Libraries**




```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

# **2. Checking Dataset**

```python
data=pd.read_csv('MentalHealthSurvey.csv')
print(data.head())
```


# *   Dataset Info



```python
data.info()
```



# *   Fixing dataset




```python
data=data.dropna(inplace=True)
```

# **Relation Between Different Factors Affecting The Mental Health Of A Student**

**1.   Relation b/w Gender vs. Depression**

```python
sns.set_theme(style="darkgrid")

gender_depression = data.groupby('gender')['depression'].mean()

sns.barplot(x=gender_depression.index, y=gender_depression.values,palette='viridis')
font1 = {'fontsize': 12}
font2 = {'fontsize': 14, 'fontweight': 'bold'}
plt.xlabel('Gender',fontdict=font1)
plt.ylabel('Average Depression Score',fontdict=font1)
plt.title('Average Depression Score by Gender',fontdict=font2)

plt.tight_layout()
plt.show()
```

**2.   Relation b/w Gender vs. Anxiety**

```python

gender_anxiety = data.groupby('gender')['anxiety'].mean()
sns.barplot(x=gender_anxiety.index, y=gender_anxiety.values, palette="magma")
plt.xlabel('Gender',font1)
plt.ylabel('Average Anxiety Score',font1)
plt.title('Average Anxiety Score by Gender',font2)

plt.tight_layout()
plt.show()
```

**3.   Relation b/w Gender vs. Financial Concerns**

```python

gender_financial_concerns = data.groupby('gender')['financial_concerns'].mean()
sns.barplot(x=gender_financial_concerns.index, y=gender_financial_concerns, palette="viridis")
plt.xlabel('Gender',font1)
plt.ylabel('Average Financial Concern Score',font1)
plt.title('Average Financial Concern Score by Gender',font2)
plt.tight_layout()
plt.show()
```

**4.   Relation b/w Gender vs. Academic Workload**

```python

gender_academic_workload = data.groupby('gender')['academic_workload '].mean()
sns.barplot(x=gender_academic_workload.index, y=gender_academic_workload.values, palette="magma")
plt.xlabel('Gender', **font1)
plt.ylabel('Average Academic Workload Score', **font1)
plt.title('Average Academic Workload Score by Gender', **font2)
plt.tight_layout()
plt.show()
```

**5.   Relation b/w Gender vs. Academic Pressure**

```python

gender_academic_pressure = data.groupby('gender')['academic_pressure'].mean()
sns.barplot(x=gender_academic_pressure.index, y=gender_academic_pressure.values, palette="viridis")
plt.xlabel('Gender', **font1)
plt.ylabel('Average Academic Pressure Score', **font1)
plt.title('Average Academic Pressure Score by Gender', **font2)
plt.tight_layout()
plt.show()
```

# **Factors Negatively Affecting Mental Health**

```python
sns.set_theme(style="whitegrid")

plt.figure(figsize=(20, 14))

#First Graph

plt.subplot(2, 2, 1)
sns.regplot(x='financial_concerns', y='depression', data=data, scatter_kws={'color':'blue'}, line_kws={'color':'red'})
plt.title('Correlation between Financial Concerns and Depression',**font2)
plt.xlabel('Financial Concerns',**font2)
plt.ylabel('Depression Score',**font2)

#Second Graph

plt.subplot(2, 2, 2)
sns.regplot(x='academic_pressure', y='depression', data=data, scatter_kws={'color':'blue'}, line_kws={'color':'red'})
plt.title('Correlation between Academic Pressure and Depression',**font2)
plt.xlabel('Academic Pressure',**font2)
plt.ylabel('Depression Score',**font2)

#Third Graph

plt.subplot(2, 2, 3)
sns.regplot(x='future_insecurity', y='depression', data=data, scatter_kws={'color':'blue'}, line_kws={'color':'red'})
plt.title('Correlation between Future Insecurity and Depression',**font2)
plt.xlabel('Future Insecurity',**font2)
plt.ylabel('Depression Score',**font2)

#Fourth Graph

plt.subplot(2, 2, 4)
sns.regplot(x='isolation', y='depression', data=data, scatter_kws={'color':'blue'}, line_kws={'color':'red'})
plt.title('Correlation between Isolation and Depression',**font2)
plt.xlabel('Isolation',**font2)
plt.ylabel('Depression Score',**font2)

plt.tight_layout()
plt.show()
```

# **Factors Positively Affecting Mental Health**


```python
plt.figure(figsize=(20, 14))
plt.subplot(2,2,1)
sns.regplot(x='social_relationships', y='depression', data=data, scatter_kws={'color':'blue'}, line_kws={'color':'green'})
plt.title('Correlation between Social Relationships and Depression',font2)
plt.xlabel('Social Relationships',font2)
plt.ylabel('Depression Score',font2)

plt.subplot(2,2,2)
sns.regplot(x='study_satisfaction', y='depression', data=data, scatter_kws={'color':'blue'}, line_kws={'color':'green'})
plt.title('Correlation between Study Satisfaction and Depression',font2)
plt.xlabel('Study Satisfaction',font2)
plt.ylabel('Depression Score',font2)


data = pd.read_csv('MentalHealthSurvey.csv')
data['stress_relief_activities_count'] = data['stress_relief_activities'].apply(lambda x: len(x.split(',')))
data.drop('stress_relief_activities', axis=1, inplace=True)
data.rename(columns={'stress_relief_activities_count': 'stress_relief_activities'}, inplace=True)

plt.subplot(2,2,3)
sns.regplot(x='stress_relief_activities', y='depression', data=data, scatter_kws={'color':'blue'}, line_kws={'color':'green'})
plt.title('Correlation between Stress Relief Activities and Depression',font2)
plt.xlabel('Stress Relief Activities',font2)
plt.ylabel('Depression Score',font2)
plt.show()

plt.tight_layout()
plt.show()
```
